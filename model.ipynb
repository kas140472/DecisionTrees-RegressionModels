### (c) Convert the decision rules into a set of IF-THEN rules.

```python
#from https://www.kdnuggets.com/2017/05/simplifying-decision-tree-interpretation-decision-rules-python.html.

def tree_to_code(tree, feature_names):

    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    print("def tree({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "  " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, threshold))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:    if {} > {}".format(indent, name, threshold))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print("{}return {}".format(indent, tree_.value[node]))

    recurse(0, 1)
    
tree_to_code(decision_tree_classifier, ['Temperature of patient','Occurrence of nausea','Lumbar pain','Urine pushing','Micturition pains','Burning of urethra, itch, swelling of urethra outlet'])
```

### (d) Use cost-complexity pruning to find a minimal decision tree and a set of decision rules with high interpretability.

```python
X_train, X_test, y_train, y_test = train_test_split( X_data, y_data, test_size=0.3, random_state=42)
print(X_train, X_test, y_train, y_test)
```

```python
dt_predict = DecisionTreeClassifier(random_state=42)

def evaluate_decision_tree(X_train, y_train, X_test, y_test):
    dt_predict.fit(X_train, y_train)
    
    y_pred = dt_predict.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    
    print('Test accuracy of the Decision Tree:', accuracy)
    print('Test error of the Decision Tree:', 1 - accuracy)

evaluate_decision_tree(X_train, y_train, X_test, y_test)
```

```python
path = dt_predict.cost_complexity_pruning_path(X_train, y_train)
alphas = path['ccp_alphas']

test_accuracy = [
    [alpha, accuracy_score(y_test, DecisionTreeClassifier(ccp_alpha=alpha).fit(X_train, y_train).predict(X_test))]
    for alpha in alphas
]

test_accuracy
```

### (f) Fit a linear model using least squares to the training set and report the test error.

```python
trainX = XY_train_imputed.iloc[:, :-1]
trainY = XY_train_imputed.iloc[:, -1]

testX = XY_test_imputed.iloc[:, :-1]
testY = XY_test_imputed.iloc[:, -1]

def calculate_test_mse(model, trainX, trainY, testX, testY):
    model.fit(trainX, trainY)
    predY = model.predict(testX)
    test_mse = mean_squared_error(testY, predY)
    return test_mse
```

```python
lr_model = LinearRegression()
lr_test_mse = calculate_test_mse(lr_model, trainX, trainY, testX, testY)

print('Test Error using Linear Regression Model:', lr_test_mse)
```

### (g) Fit a ridge regression model on the training set, with λ chosen by cross-validation. Report the test error obtained.

```python
def calculate_cv_test_mse(model, trainX, trainY, testX, testY, alphas):
    cv_split = KFold(5, shuffle=True, random_state=7)
    if model=='RidgeCV':
        model = RidgeCV(alphas=alphas, cv=cv_split)
    elif model=='LassoCV':
        model = LassoCV(alphas=alphas, cv=cv_split)
    
    model.fit(trainX, trainY)
    print("Alpha selected = {}".format(model.alpha_))
    predY = model.predict(testX)
    test_mse = mean_squared_error(testY, predY)
    return model, test_mse
```

```python
alphas = [10 ** i for i in range(-5, 5)]
rr_model, rr_test_mse = calculate_cv_test_mse('RidgeCV', trainX, trainY, testX, testY, alphas)

print('Test Error using Ridge Regression Model:', rr_test_mse)
```

### (h) Fit a LASSO model on the training set, with λ chosen by cross-validation. Report the test error obtained, along with a list of the variables selected by the model. Repeat with standardized features. Report the test error for both cases and compare them.

```python
alphas = [10 ** i for i in range(-5, 5)]
ls_model, ls_test_mse = calculate_cv_test_mse('LassoCV', trainX, trainY, testX, testY, alphas)

print('Test Error using Lasso Model:', ls_test_mse)
print("\n")
ls_selected_features = [testX.columns[x] for x in np.where(ls_model.coef_ != 0)[0]]
print("The ", len(ls_selected_features), " variables selected by the Lasso model are:")
print("\n")
print(ls_selected_features)
```

```python
scale = StandardScaler()

trainX_standard = scale.fit_transform(trainX)
testX_standard = scale.fit_transform(testX)

trainX_standard = pd.DataFrame(data=trainX_standard, columns=trainX.columns)
testX_standard = pd.DataFrame(data=testX_standard, columns=testX.columns)

ls_model_standard, ls_test_mse_standard = calculate_cv_test_mse('LassoCV', trainX_standard, trainY, testX_standard, testY, alphas)

print('Test Error using Lasso Model (with standardized features):', ls_test_mse_standard)
print("\n")

ls_selected_features_standard = [testX_standard.columns[x] for x in np.where(ls_model_standard.coef_ != 0)[0]]
print("The ", len(ls_selected_features_standard), " variables selected by the Lasso model (with standardized features) are:")
print("\n")
print(ls_selected_features_standard)
```

Comparing test errors:

### (i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
def calculate_pca(trainX_pca, n_components):
    result = PCA(n_components=n_components).fit(trainX_pca)
    return result

def PCR_model(trainX_pca, trainY_pca, f, dim):
    pca_val_error = 0
    cv_split = KFold(f, shuffle=True, random_state=7)
    
    for train_idx, val_idx in cv_split.split(trainX_pca):
        trainX_temp, trainY_temp = trainX_pca.iloc[train_idx, :], trainY_pca.iloc[train_idx]
        valX_temp, valY_temp = trainX_pca.iloc[val_idx, :], trainY_pca.iloc[val_idx]
        
        train_pca = calculate_pca(trainX_temp, dim).transform(trainX_temp)
        val_pca = calculate_pca(trainX_temp, dim).transform(valX_temp)
        
        lr_model_pcr = LinearRegression()
        
        pca_val_error += calculate_test_mse(lr_model_pcr, train_pca, trainY_temp, val_pca, valY_temp)
        
    return pca_val_error / f

avg_pca_val_error = [(M, PCR_model(trainX, trainY, 5, M)) for M in range(1, len(trainX.columns) + 1)]
avg_pca_val_error.sort(key=lambda x: x[1])

min_val_mse, best_pca_components = avg_pca_val_error[0][1], avg_pca_val_error[0][0]
print('Minimum validation MSE is {} when M is {}'.format(min_val_mse, best_pca_components))

pca_train_res = calculate_pca(trainX, best_pca_components).transform(trainX)
pca_test_res = calculate_pca(trainX, best_pca_components).transform(testX)

lr_model_pcr = LinearRegression()
pcr_test_mse = calculate_test_mse(lr_model_pcr, pca_train_res, trainY, pca_test_res, testY)

print("Test error using PCR model:", pcr_test_mse)
```

### (j) In this section, we would like to fit a boosting tree to the data. As in classification trees, one can use any type of regression at each node to build a multivariate regression tree. Because the number of variables is large in this problem, one can use L1-penalized regression at each node. Such a tree is called L1 penalized gradient boosting tree. You can use XGBoost to fit the model tree. Determine α (the regularization term) using cross-validation.

```python
parameters = {
    'estimator': xgb.XGBRegressor(n_estimators=100, max_depth=4),
    'param_grid': {'alpha': alphas},
    'cv': KFold(5, shuffle=True, random_state=7)
}

xgb_cv_model = GridSearchCV(**parameters)
xgb_cv_model.fit(trainX, trainY)

xgb_optimum_alpha = xgb_cv_model.best_params_['alpha']
print("Best alpha using cross-validation:", xgb_optimum_alpha)

xgb_model = xgb.XGBRegressor(reg_alpha=xgb_optimum_alpha)

xgb_test_mse = calculate_test_mse(xgb_model, trainX, trainY, testX, testY)
print("Test error using XGBoost model:", xgb_test_mse)
```
